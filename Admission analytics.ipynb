```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
# 1. Patients frequency of visit
file_path = 'admissions.csv'
df = pd.read_csv(file_path)
grouped_data = df.groupby('subject_id').count() 
grouped_data = grouped_data.reset_index()
grouped_data = grouped_data.rename(columns={'hadm_id': 'frequency'})
selected_col = grouped_data[['subject_id', 'frequency']]
```

```python
# Save the patient frequency to excel
selected_col.to_excel('Patitent_frequency.xlsx',sheet_name='Data' )
```

```python
total_fre = selected_col.frequency.value_counts()
total_fre
```

```python
total_fre.to_excel('Patitent_unique_frequency_count.xlsx',sheet_name='Data' )
```

```python
# Plot the histogram for the number of visits
plt.hist(selected_col['frequency'],range=(1,100), color='blue') 
plt.xlabel('Frequency_of_visit')
plt.ylabel('Frequency')
plt.title('Bar Chart of Subject ID Frequencies')
plt.show()
```

```python
#2. Categorizing patients based on Admission type

file_path = 'admissions.csv'
df = pd.read_csv(file_path)
grouped_data = df.groupby('admission_type').count() 
grouped_data = grouped_data.reset_index()
grouped_data = grouped_data.rename(columns={'hadm_id': 'frequency'})
Admission_type = grouped_data[['admission_type', 'frequency']]
```

```python
Admission_type.to_excel('Admission_type.xlsx',sheet_name='Data' )
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(Admission_type['admission_type'], Admission_type['frequency'], color='skyblue')
plt.xlabel('Admission Type')
plt.ylabel('Frequency')
plt.title('Admission Type Frequency')
plt.xticks(rotation=90, ha='right')  
plt.tight_layout()
```
